# Preprocessing

### 1. Load the dataset and the model

```python
!pip install datasets peft transformers huggingface_hub

from peft import PeftModel, PeftConfig
from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset
from transformers import AutoTokenizer

notebook_login()

# Load the tokenizer from the base Llama-2 7B Chat model
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat")

# Load the configuration and model from Hugging Face
config = PeftConfig.from_pretrained("AIRLab-POLIMI/llama-2-7b-chat-hf-btgenbot-adapter")
base_model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf")
model = PeftModel.from_pretrained(base_model, "AIRLab-POLIMI/llama-2-7b-chat-hf-btgenbot-adapter")

# Load the BTGenBot dataset from Hugging Face
ds = load_dataset("AIRLab-POLIMI/btgenbot")
```

### 2. Preprocess btgenbot dataset

```python
# Preprocess the dataset (tokenizing the input)
def preprocess_data(example):
    return tokenizer(example['input'], truncation=True, padding="max_length")

tokenized_data = ds.map(preprocess_data, batched=True)

# Convert the tokenized dataset to PyTorch tensors
tokenized_data.set_format(type='torch', columns=['input_ids', 'attention_mask'])
```
